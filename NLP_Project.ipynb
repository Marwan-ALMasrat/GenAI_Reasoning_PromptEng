```python
!pip install spacy contractions
!python -m spacy download en_core_web_sm

import spacy
import contractions
import string

nlp = spacy.load("en_core_web_sm")

def clean_text_spacy(text):
    text = contractions.fix(text)
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    doc = nlp(text)
    cleaned_text = " ".join([token.text for token in doc if not token.is_stop])
    return cleaned_text

df['clean_text'] = df['clean_text'].apply(clean_text_spacy)
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import numpy as np
import torch.nn.functional as F

# 1) تحميل النموذج
model_name = "cardiffnlp/twitter-roberta-base-sentiment-latest"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# نقل النموذج إلى GPU (يستخدم المتغير device من الكود السابق)
model = model.to(device)
print(f"✅ النموذج تم تحميله على: {device}")

# 2) دالة تحليل المشاعر لنص واحد
def get_sentiment(text):
    # التأكد من أن النص ليس فارغاً أو NaN
    if pd.isna(text) or text == "":
        return "neutral", 0.0

    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    # نقل المدخلات إلى GPU
    inputs = {k: v.to(device) for k, v in inputs.items()}

    outputs = model(**inputs)
    scores = F.softmax(outputs.logits, dim=1).detach().cpu().numpy()[0]
    labels = ["negative", "neutral", "positive"]
    sentiment = labels[np.argmax(scores)]
    score = float(np.max(scores))
    return sentiment, score

# 3) تطبيق الدالة على DataFrame (df يجب أن يحتوي عامود text)
print("\n🔄 بدء تحليل المشاعر...")
df['sentiment'], df['sentiment_score'] = zip(*df['clean_text'].apply(get_sentiment))
print("✅ تم الانتهاء من تحليل المشاعر!")

# 4) إظهار أول النتائج
print("\n📊 عينة من النتائج:")
print(df[['clean_text', 'sentiment', 'sentiment_score']].head(10))
```

###Sentiment vs. Rating Mismatch Analysis

```python
high_negative = df[(df['stars'] >= 4) & (df['sentiment'] == 'negative')]
print("High stars but negative:", len(high_negative))

low_positive = df[(df['stars'] <= 2) & (df['sentiment'] == 'positive')]
print("Low stars but positive:", len(low_positive))

print("Total:", len(high_negative) + len(low_positive))
```

```python
df = df[~((df['stars'] >= 4) & (df['sentiment'] == 'negative'))]

df = df[~((df['stars'] <= 2) & (df['sentiment'] == 'positive'))]
```
